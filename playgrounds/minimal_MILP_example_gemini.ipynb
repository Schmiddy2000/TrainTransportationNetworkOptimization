```python
import gurobipy as gp
from gurobipy import GRB
```

```python
# 1. Initialize the Model
model = gp.Model("Basic_MILP")
```

```python
# 2. Define Decision Variables
# Syntaxes: model.addVar() for a single variable, model.addVars() for multiple
x = model.addVar(vtype=GRB.CONTINUOUS, name="x", lb=0)  # Continuous, >= 0
y = model.addVar(vtype=GRB.BINARY, name="y")          # Binary {0, 1}
z = model.addVar(vtype=GRB.INTEGER, name="z", lb=1, ub=10) # Integer between 1 and 10
```

```python
# 3. Set the Objective Function
# Let's say we want to maximize: 2x + 3y - z
model.setObjective(2*x + 3*y - z, GRB.MAXIMIZE)
```

```python
# 4. Add Constraints
# Constraint 1: x + 2y <= 4
model.addConstr(x + 2*y <= 4, name="c1")

# Constraint 2: x + z == 5
model.addConstr(x + z == 5, name="c2")
```

```python
# 5. Optimize the Model
model.optimize()
```

```python
# 6. Retrieve the Results
if model.Status == GRB.OPTIMAL:
    print("\n--- Optimal Solution Found ---")
    print(f"Objective Value: {model.ObjVal}")
    print(f"x = {x.X}")
    print(f"y = {y.X}")
    print(f"z = {z.X}")
else:
    print("Optimization was not successful.")
```
